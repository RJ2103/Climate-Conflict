```python
temp = pd.read_csv(r"C:\Users\Admin\Desktop\TheObservable\Climate-Conflict\data\processed\temperature_zonal_annual.csv")
conflict = pd.read_csv(r"C:\Users\Admin\Desktop\TheObservable\Climate-Conflict\data\processed\conflict_country_year_latband.csv")

temp.head()
```

```python
conflict.shape
```

```python
conflict_band_year = (
    conflict
    .groupby(["year", "latitude"])
    .agg(conflict_countries = ("conflict_present", "sum"))
    .reset_index()
    )
```

```python
pd.set_option('display.max_rows', 50)
conflict_band_year.head()
```

```python
# See what each group contains
for name, group in conflict_band_year:
    year, lat_band = name
    print(f"\n{year} - {lat_band}:")
    print(f"  Countries: {list(group['country'])}")
    print(f"  Total conflicts: {group['conflict_present'].sum()}")
    print(f"  Total deaths: {group['deaths'].sum()}")
```
